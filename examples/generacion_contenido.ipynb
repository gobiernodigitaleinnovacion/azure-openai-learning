```python
# -*- coding: utf-8 -*-
import sys
import os

# Configuración de ruta para importar el generador de contenido
current_dir = os.getcwd()  # Obtiene el directorio actual
sys.path.append(os.path.dirname(os.path.dirname(current_dir)))

from src.generador_contenido import GeneradorContenido

# Inicializa el generador de contenido
generador = GeneradorContenido()

# Función para guardar contenido en un archivo
def guardar_en_archivo(nombre_archivo, contenido):
    with open(nombre_archivo, "w", encoding="utf-8") as archivo:
        archivo.write(contenido)
    print(f"Contenido guardado en {nombre_archivo}")

# Generar Email
contexto_email = "Solicitud de reunión para discutir el nuevo proyecto de IA"
email = generador.generar_email(contexto_email, tipo="formal")
print("EMAIL GENERADO\n", email, "\n")
guardar_en_archivo("email_generado.txt", email)

# Generar Blog
tema_blog = "Impacto de la Inteligencia Artificial en la Medicina Moderna"
blog = generador.crear_blog(tema_blog, longitud="media")
print("ARTÍCULO DE BLOG\n", blog, "\n")
guardar_en_archivo("blog_generado.txt", blog)

# Generar Descripción de Producto
producto = "Smartwatch XYZ"
caracteristicas = """
- Pantalla AMOLED de 1.4"
- Batería de 7 días
- Resistente al agua
- Monitores de salud
- GPS integrado
"""
descripcion = generador.describir_producto(producto, caracteristicas)
print("DESCRIPCIÓN DE PRODUCTO\n", descripcion, "\n")
guardar_en_archivo("descripcion_producto.txt", descripcion)

# Generar Informe Técnico
datos_informe = """
Métricas del Proyecto Q4 2024:
- Tiempo de desarrollo: 3 meses
- Recursos utilizados: 5 desarrolladores
- Presupuesto ejecutado: 80%
- Objetivos completados: 90%
"""
informe = generador.generar_informe(datos_informe, tipo_informe="ejecutivo")
print("INFORME TÉCNICO\n", informe, "\n")
guardar_en_archivo("informe_tecnico.txt", informe)

```
