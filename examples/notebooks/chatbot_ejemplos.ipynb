```python
# examples/chatbot_ejemplos.ipynb

import sys
from pathlib import Path
current_dir = Path.cwd()
parent_dir = current_dir.parent
sys.path.append(str(parent_dir))

from src.chatbot_ia import ChatbotIA

# Crear instancia del chatbot
chatbot = ChatbotIA()

# Ejemplo de soporte técnico
print("1. Soporte Técnico:")
preguntas_soporte = [
    "No puedo acceder a mi cuenta",
    "¿Cómo cambio mi contraseña?",
    "El sistema está muy lento"
]

for pregunta in preguntas_soporte:
    respuesta = chatbot.responder(pregunta, "soporte técnico")
    print(f"\nUsuario: {pregunta}")
    print(f"Chatbot: {respuesta}")

# Limpiar historial para nuevo contexto
chatbot.limpiar_historial()

# Ejemplo de ventas
print("\n2. Consultas de Ventas:")
preguntas_ventas = [
    "¿Qué planes tienen disponibles?",
    "¿Hay descuentos para empresas?",
    "¿Puedo hacer una prueba gratuita?"
]

for pregunta in preguntas_ventas:
    respuesta = chatbot.responder(pregunta, "ventas")
    print(f"\nUsuario: {pregunta}")
    print(f"Chatbot: {respuesta}")
```
